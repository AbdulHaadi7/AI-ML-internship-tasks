# 1. Importing Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn plots
sns.set(style="whitegrid")
```

# 2. Loading the Dataset with Pandas

```python
# Load dataset from a local CSV file
df = pd.read_csv("/content/Iris.csv")  # Make sure the path is correct
```

```python
# Display shape
print("Shape of dataset:", df.shape)

# Column names
print("Column names:", df.columns.tolist())

# First five rows
print("\nFirst 5 rows:")
print(df.head())
```

# 4. Summary Statistics

```python
# Info about data types and non-null values
df.info()

# Descriptive statistics
print("\nSummary statistics:")
print(df.describe())
```

# 5. Visualization

## 1. Scatter Plot

```python
sns.pairplot(df, hue='Species', height=2.5)
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

## 2. Histogram

```python
df.hist(figsize=(10, 6), color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```

## 3. Box Plots

```python
plt.figure(figsize=(15, 12))

for i, column in enumerate(df.select_dtypes(include='number').columns, 1):
    plt.subplot(3, 2, i)  # 3 rows, 2 columns = 6 slots
    sns.boxplot(x='Species', y=column, data=df)
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()

```
